```python
import pandas as pd
import json

with open('kul_professor.json', 'r', encoding='utf-8') as f:
    kul_professor = json.load(f)

df = pd.DataFrame(kul_professor)
df
```

```python
df = df.rename(columns={'curriculum_link': 'orcid_link'})
df.head()
```

```python
df['photo_url'] = df['photo_url'].replace('None', None)
df['orcid_link'] = df['orcid_link'].replace('None', None)
df.head()
```

```python
df.to_json('kul_professors.json', orient='records', force_ascii=False, indent=2)
```
