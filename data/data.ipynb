```python
import json
from collections import Counter, defaultdict

# Load data from professors.json
with open('professors.json', 'r', encoding='utf-8') as f:
    professors = json.load(f)

# Number of professors by university
professors_by_university = defaultdict(int)
names = []
units = []

for prof in professors:
    university = prof.get('university')
    name = prof.get('name')
    unit = prof.get('unit')
    if university:
        professors_by_university[university] += 1
    if name:
        names.append(name)
    if unit:
        units.append(unit)

# Most used names (first names)
first_names = [n.split()[0] for n in names if n]
most_common_names = Counter(first_names).most_common(10)

# Most used units
# Flatten the units list (some entries are lists)
flat_units = []
for u in units:
    if isinstance(u, list):
        flat_units.extend(u)
    elif u:
        flat_units.append(u)
most_common_units = Counter(flat_units).most_common(10)

# Results
professors_by_university = dict(professors_by_university)
most_common_names = dict(most_common_names)
most_common_units = dict(most_common_units)

print("Professors by university:", professors_by_university)
print("Most used names:", most_common_names)
print("Most used units:", most_common_units)
```

```python
import matplotlib.pyplot as plt

# Professors by university
plt.figure(figsize=(10, 5))
plt.bar(professors_by_university.keys(), professors_by_university.values())
plt.title('Number of Professors by University')
plt.xlabel('University')
plt.ylabel('Number of Professors')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Most used names
plt.figure(figsize=(8, 4))
plt.bar(most_common_names.keys(), most_common_names.values(), color='orange')
plt.title('Top 10 Most Common First Names')
plt.xlabel('First Name')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Most used units
plt.figure(figsize=(10, 5))
plt.bar(most_common_units.keys(), most_common_units.values(), color='green')
plt.title('Top 10 Most Common Units')
plt.xlabel('Unit')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
